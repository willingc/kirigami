```python
from pydiscourse import DiscourseClient
client = DiscourseClient(
        'https://discuss.python.org',
        api_username='willingc',
        api_key = None
)
```

```python
client.user('willingc')
```

```python
posts = client.topic_posts(46593)
```

```python
posts
```

```python
type(posts)
```

```python
import json


# Convert Python to JSON  
json_object = json.dumps(posts, indent = 4) 
 
# Print JSON object
print(json_object) 
```

```python
import io
import pandas as pd

df = pd.read_json(io.StringIO(json_object))
```

```python
df.head()
```

```python
df_normalized = pd.json_normalize(df['post_stream'])

df_normalized.head()
```

```python
df_normalized_transposed = df_normalized.T
df_normalized_transposed
```
